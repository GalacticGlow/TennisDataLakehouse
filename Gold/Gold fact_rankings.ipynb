```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

#Loading from silver layer

```python
df = spark.table('tennislakehouse.silver.rankings')
```

#Writing to gold layer with new id

```python
w = Window.orderBy(
    'ranking_date',
    'player_id'
)

df = df.withColumn(
    'record_id',
    F.row_number().over(w)
)

df = df.select(
    "record_id",
    *[c for c in df.columns if c != "record_id"]
)

(
df.write
    .mode('overwrite')
    .format('delta')
    .option('mergeSchema', 'true')
    .saveAsTable('tennislakehouse.gold.fact_rankings')
)
```
