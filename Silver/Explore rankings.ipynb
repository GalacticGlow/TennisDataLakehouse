#Exploring the table Rankings

```sql
SELECT
  *
FROM tennislakehouse.bronze.rankings
LIMIT 100
```

```sql
SELECT
  *
FROM (
  SELECT 
    *,
    ROW_NUMBER() OVER ( PARTITION BY player, rank, ranking_date ORDER BY player, rank, ranking_date ) flag
  FROM tennislakehouse.bronze.rankings
)
WHERE flag != 1
```

```sql
SELECT
  *
FROM tennislakehouse.bronze.rankings
WHERE ranking_date = 20210719 AND player = 106254 AND rank = 740
```

```sql
SELECT
  *
FROM tennislakehouse.bronze.rankings
WHERE player = 104925
ORDER BY ranking_date
```

```sql
--no invalid dates found
SELECT
  *
FROM tennislakehouse.bronze.rankings
WHERE TO_DATE(CAST(ranking_date AS STRING), 'yyyyMMdd') IS NULL
```
