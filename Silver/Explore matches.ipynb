#Verifying allowed values in the columns

```sql
SELECT
  COUNT(surface)
FROM tennislakehouse.bronze.matches
WHERE surface = 'Carpet'
```

```sql
SELECT
  COUNT(*)
FROM tennislakehouse.bronze.matches
WHERE draw_size NOT IN ('2', '4', '8', '16', '32', '64', '128')
```

```sql
SELECT
  COUNT(*)
FROM tennislakehouse.bronze.matches
WHERE LEN(tourney_date) != 8
```

```sql
SELECT
  COUNT(match_num)
FROM tennislakehouse.bronze.matches
WHERE TRY_CAST(match_num AS INT) <= 127
```

```sql
SELECT
  COUNT(winner_id)
FROM tennislakehouse.bronze.matches
WHERE LEN(winner_id) != 6
```

```sql
SELECT
  COUNT(loser_id)
FROM tennislakehouse.bronze.matches
WHERE LEN(loser_id) != 6;
```

```sql
SELECT DISTINCT
  match_num
FROM tennislakehouse.bronze.matches
```

```sql
SELECT
  *
FROM tennislakehouse.bronze.matches
WHERE TRY_CAST(loser_seed AS INT) > 32
```

```sql
SELECT DISTINCT
  loser_age
FROM tennislakehouse.bronze.matches
ORDER BY loser_age DESC
```

```sql
SELECT DISTINCT
  score
FROM tennislakehouse.bronze.matches
WHERE LEN(score) < 7
```

```sql
SELECT DISTINCT
  best_of
FROM tennislakehouse.bronze.matches
WHERE LEN(best_of) = 1
```

```sql
SELECT
  LEN(best_of)
FROM tennislakehouse.bronze.matches
WHERE best_of = '1'
```

#Allowed values (for some columns):
-surface: Hard, Grass, Clay, Carpet

-draw_size: 2, 4, 8, 16, 32, 64, 128

-match_num: has to be <= 127

-winner_seed and loser_seed: has to be between 1 and 32, both inclusive

-score: has to be either longer than 7 characters or be '6-7' or similar

-best_of: 1, 3, 5

-round: R128, R64, R32, R16, QF, SF, F

-minutes: has to be <= 665, as the longest tennis match in history took 665 minutes
- tourney_level: 
| Code                             | Meaning                                             |
| -------------------------------- | --------------------------------------------------- |
| **G**                            | Grand Slam                                          |
| **M**                            | Masters 1000                                        |
| **A**                            | ATP 500 / ATP 250 (regular tour events)             |
| **C**                            | Challenger                                          |
| **F**                            | Futures / ITF tournaments                           |
| **D**                            | Davis Cup                                           |
| **O**                            | Olympics                                            |
| **S**                            | Possibly “Special” / older tour events / satellites |
| **R**                            | Round-robin events (e.g. ATP Finals group stage)    |
| **L**                            | Laver Cup                                           |
| **U**                            | Unknown (to be replaced with 'N/A')                                           |
- winner_entry and loser_entry:
| Code               | Meaning                                                    |
| ------------------ | ---------------------------------------------------------- |
| **Q**              | Qualifier (came through qualifying rounds)                 |
| **WC**             | Wildcard (invited by organizers)                           |
| **LL**             | Lucky Loser (lost in qualifying but replaced someone)      |
| **ALT / Alt / AL** | Alternate (backup player who gets in if someone withdraws) |
| **PR**             | Protected Ranking (injury protection ranking)              |
| **SE**             | Special Exempt (still playing previous tournament)         |
| **JE / JR**        | Junior Exempt (top junior allowed into pro event)          |
| **ITF**            | ITF entry (lower-tier / federation pathway)                |
| **IR**             | Injury Replacement                                         |
| **A**              | Direct Acceptance (normal entry via ranking)               |
